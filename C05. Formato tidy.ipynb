Ahora bien, en caso de estar en la situación de tener una tabla en formato ancho nos va a interesar pasar a formato largo.

Contamos con las funciones ``melt`` y ``pd.melt`` que tienen la siguiente sintaxis:

```r
################################################################
#######       En R       #######################################
################################################################

#install.packages("reshape2")
library(reshape2)

melt(data=tabla, #es la tabla no ordenada
     id.vars=c("col1","col2",...),  #columnas que quieres mantener
     measure.vars=c("col3","col4",...), #columnas que quieres ordenar
     variable.name="variable", #nombre de la nueva columna donde se colocarán los nombres de las columnas antiguas   
     value.name="value") #nombre de la nueva columna donde se colocarán los valores correspondientes a las columnas antiguas   
     
```

```python
################################################################
#######    En Python     #######################################
################################################################

import pandas as pd

melt(frame=tabla, #es la tabla no ordenada
     id_vars=["col1","col2",...],  #columnas que quieres mantener
     value_vars=["col3","col4",...], #columnas que quieres ordenar
     var_name="variable", #nombre de la nueva columna donde se colocarán los nombres de las columnas antiguas   
     value_name="value") #nombre de la nueva columna donde se colocarán los valores correspondientes a las columnas antiguas   
```

```r
################################################################
#######       En R       #######################################
################################################################

#install.packages("reshape2")
library(reshape2)

medallero_tidy = melt(data=medallero_no_tidy, 
     id.vars=c("pais","anio"),
     measure.vars=c("hombres","mujeres"), 
     variable.name="sexo",   
     value.name="medallas")
```

```python
################################################################
#######    En Python     #######################################
################################################################

import pandas as pd

medallero_tidy = pd.melt(frame=medallero_no_tidy, 
                         id_vars=["pais", "anio"], 
                         value_vars=["hombres", "mujeres"], 
                         var_name="sexo", 
                         value_name="medallas")
```
